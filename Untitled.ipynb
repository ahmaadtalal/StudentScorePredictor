```python
features_top5 = ["Attendance", "Hours_Studied", "Previous_Scores", 
                 "Access_to_Resources", "Parental_Involvement"]

features_academic = ["Attendance", "Hours_Studied", "Previous_Scores", "Tutoring_Sessions"]

model = LinearRegression()

def evaluate_features(X, y, feature_list, name):
    X_subset = X[feature_list]
    scores = cross_val_score(model, X_subset, y, cv=5, scoring="r2")
    mse_scores = -cross_val_score(model, X_subset, y, cv=5, scoring="neg_mean_squared_error")
    
    print(f"\n{name}")
    print("Cross-Validation R² Scores:", scores)
    print("Mean R²:", np.mean(scores))
    print("Std Dev R²:", np.std(scores))
    print("Mean MSE:", np.mean(mse_scores))

evaluate_features(X, y, features_top5, "Top 5 Correlated Features")
evaluate_features(X, y, features_academic, "All Academic-Related Features")
```

```python
pruned_features = [
    col for col in X.columns 
    if col not in ["Gender", "School_Type", "Sleep_Hours", 
                   "Physical_Activity", "Learning_Disabilities", "Distance_from_Home"]
]

X_pruned = X[pruned_features]

r2_scores = cross_val_score(model, X_pruned, y, cv=5, scoring="r2")
mse_scores = -cross_val_score(model, X_pruned, y, cv=5, scoring="neg_mean_squared_error")

print("Pruned Feature Set")
print("Cross-Validation R² Scores:", r2_scores)
print("Mean R²:", np.mean(r2_scores))
print("Std Dev R²:", np.std(r2_scores))
print("Mean MSE:", np.mean(mse_scores))
```
